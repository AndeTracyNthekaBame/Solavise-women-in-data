```python
import math

""" Calculating area of a circle """
print("Welcome. I am your calculator. Play along")
radius = float(input("Please enter the radius of a cirle"))
area_circle = math.pi * ( radius**2)
print(f"\n The area of your circle with radius {radius} is {area_circle:.2f}  square units. \n")

""" Calculating area of a Triangle """
base = float(input("Please enter the base of your triangle"))
height = float(input("Please enter the Height of your triangle"))
area_triangle = 0.5 * base * height
print(f"\n The area of the triangle with base {base} and height {height} is {area_triangle:.2f} square units.\n")

""" Summary """
print(f"In summary, the circle has an area of {area_circle:.2f} square units, "
      f"while the triangle has an area of {area_triangle:.2f} square units. "
      "Both results were calculated based on your input values.")
```
